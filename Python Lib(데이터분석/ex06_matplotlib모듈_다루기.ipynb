```python
# 데이터 불러와서 economi 변수에 담아보기
# 위에서 행 5개 출력해보기
economi = pd.read_csv('.\data\시도_성별_경제활동인구_총괄_KOSIS_2022.csv',encoding ='euc-kr')
economi

economi.iloc[:5]
```

```python
# 데이터 전체 속성 정보 확인하기
economi.describe()
economi.info()
# 행34, 열10
# 고용율,실업률 컬럼 결측치 발생

# 결측치 개수를 확인하는 방법
economi.isnull().sum()

# 고용률 : 생산가능연령인 15세 이상인 인구중에서 특정 시점에 취업하고 있는 사람의 비율
# 실업률 : 경제활동참가자 중에서 실업상태에 있는 사람의 비율
```

```python
# 고용률 결측치 처리해보기
# 특성공학 : 기존의 컬럼 활용해서 데이터 생성
# 고용률 : 취업자 컬럼/ 15세 이상인구 컬럼

취업자 = economi.iloc[:,4]
_15세이상 = economi.iloc[:,2]

고용률 = round((취업자/_15세이상)*100, 1)
economi['고용률(%)'] = 고용률

economi.isnull().sum()
```

```python
# 실업률 처리해보기
# 실업률 => 실업자/경제활동인구

실업자 = economi.loc[:,'실업자 (천명)']
경제활동인구 = economi.loc[:,"비경제활동인구 (천명)"]

실업률 = round((실업자/경제활동인구)*100, 1)
economi['실업률(%)'] = 실업률
display(economi)
```

```python
# 시도별 고용률 평균 구해보기
# 시도별 고용률 평균이 가장 높은 지역 5개 확인하기
# 1. 필요한 컬럼들은
# 2. groupby()
# 3. mean()
# 4. 상위 5개 데이터만 출력해보기

economi.columns # 컬럼확인


평균고용률 = economi[['행정구역(시도)','고용률(%)']].groupby('행정구역(시도)').mean().sort_values(by='고용률(%)'
,ascending=False)

# 선생님 코드
평균고용률 = economi[['행정구역(시도)','고용률(%)']].groupby('행정구역(시도)').agg(고용률평균=('고용률(%)','mean')).sort_values(by='고용률평균'
,ascending=False)
평균고용률.iloc[0:5,]

평균고용률
```

```python
y_행정 = 평균고용률['고용률평균']  # 1차원

x_행정 = 평균고용률.index

plt.figure(figsize = (6,3))
plt.bar(x_행정, y_행정)
plt.xticks(rotation=45)
plt.ylim(55,80)
plt.title('시도별 고용률 평균 시각화')
plt.xlabel('행정구역(시도)')
plt.ylabel('고용률 평균(%)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=x_행정,y=y_행정) # 색상 파레트 설정이 조금 더 다양함
plt.xticks(rotation=45)
plt.show()
```

```python
# 남자의 데이터 접근
# plt.ylim(55,80)
남자실업률 = economi.iloc[::2,[9]]
# y_남자 = 남자실업률['실업률(%)']
# display(economi)
y_남자 = economi.query('성별 == "남자"')['실업률(%)']
x_지역 = economi['행정구역(시도)']


# 선생님 방법
eco_m_unemp = economi.loc[economi['성별'] == '남자',['행정구역(시도)','실업률(%)']]

x_지역 = eco_m_unemp['행정구역(시도)']
y_남자 = eco_m_unemp['실업률(%)']


sns.barplot(x=x_지역,y=y_남자)
plt.xticks(rotation=45)
plt.show()
```

#### Scatter plot(산점도 그래프), Pie plot


```python
# 산점도 그래프 그려보기
# 산점도 점 == 버블 (버블 사이즈, 버블 색상 설정)
x = np.arange(1,4) #1,2,3
np.random.seed(2)
y = np.random.rand(3) #실수 난수 생성
display(x)
display(y)

plt.scatter(x,y, s = [100,100,300], c = ['r','r','c'])
plt.show()
```

```python
# 산점도 색상bar 설정
plt.scatter(x,y, s = [100,100,300], c = range(3), cmap = 'cool')
# cmap : matplotlib 색상 파레트 설정값
plt.colorbar()
plt.show()
```

```python
# Pie plot
display(x)
display(y)
# % 구하기
for i in y:
    print(f'{i/y.sum()*100:.2f}')

plt.pie(y, labels = x, autopct='%.2f%%') # 퍼센트로 소수점 2자리까지 %로 표현
plt.show()
```
