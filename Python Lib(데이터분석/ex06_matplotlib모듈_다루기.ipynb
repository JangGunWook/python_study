#### Scatter plot( ), Pie plot


```python
#   
#   ==  ( ,   )
x = np.arange(1,4) #1,2,3
np.random.seed(2)
y = np.random.rand(3) #  
display(x)
display(y)

plt.scatter(x,y, s = [100,100,300], c = ['r','r','c'])
plt.show()
```

```python
#  bar 
plt.scatter(x,y, s = [100,100,300], c = range(3), cmap = 'cool')
# cmap : matplotlib   
plt.colorbar()
plt.show()
```

```python
# Pie plot
display(x)
display(y)
# % 
for i in y:
    print(f'{i/y.sum()*100:.2f}')

plt.pie(y, labels = x, autopct='%.2f%%') #   2 % 
plt.show()
```
