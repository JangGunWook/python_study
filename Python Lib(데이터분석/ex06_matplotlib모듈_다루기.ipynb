```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# Line Plot 선 그래프

```python
# 배열 생성하여 데이터 준비
x = np.arange(1,6)
display(x)
np.random.seed(3)
y = np.random.randint(1,10,size =5)
display(y)
```

```python
# 시각화
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker='^',c='green', ls='-.',
         lw =3, ms = 8, mec = 'red', mfc = 'red', mew = 5)
plt.show()
```

- show() 사용하기

```python
# 여러 그래프를 한 장에 그리기
# plt.show() : 도화지(그래프창)을 구분하는 역할
y2 = np.random.randint(1,6, size=5)
y3 = np.random.randint(1,6, size=5)
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
plt.plot(x,y3)

plt.show()
```

- 그래프창 크기 조절하기

```python
# plt.figure(figsize = (x크기,y크기))
plt.figure(figsize = (8,2))
plt.plot(x,y)
plt.show()
```

- 한 창에 칸 나눠서 그래프 여러개 표시하기 : subplot()

```python
plt.figure(figsize = (15,3))
plt.subplot(1,3,1) #1행 3열 의 1번째
plt.title('<1>') #그래프 제목 달기
plt.xlabel('x axis') # x축 label
plt.ylabel('y axis') # y축 label
plt.plot(x,y)

plt.subplot(1,3,2) #1행 3열 의 2번째
plt.title('<2>') #그래프 제목 달기
plt.plot(x,y2)

plt.subplot(1,3,3) #1행 3열 의 3번째
plt.title('<3>') #그래프 제목 달기
plt.plot(x,y3)

plt.show()
```

- 예제1 ) 국가통계포털의 장래_인구변동_데이터를 활용하여 lineplot으로 출생아수, 사망자수 시각화 해보기

```python
# 1. 데이터 불러오기
# 단, 인코딩 'euc-kr', index_col = 
data = pd.read_csv('.\data\장래_인구변동_KOSIS.csv', encoding ='euc-kr',
                   index_col = '인구변동요인별')
data
```

```python
# 출생아수(천명) 데이터에 접근해보기
y1 = data.iloc[1]
# 사망자수(천명) 데이터에 접근해보기
y2 = data.iloc[2]
```

```python
# 출생아수, 사망자수 선그래프 시각화
x = data.columns #y2.index

plt.figure(figsize=(4,3))
plt.plot(x,y1,marker='*')
plt.plot(x,y2,marker='o')
plt.show()
```
