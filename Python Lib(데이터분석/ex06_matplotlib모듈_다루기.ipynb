# Line Plot  

```python
#    
x = np.arange(1,6)
display(x)
np.random.seed(3)
y = np.random.randint(1,10,size =5)
display(y)
```

```python
# 
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker='^',c='green', ls='-.',
         lw =3, ms = 8, mec = 'red', mfc = 'red', mew = 5)
plt.show()
```

```python
plt.figure(figsize = (15,3))
plt.subplot(1,3,1) #1 3  1
plt.title('<1>') #  
plt.xlabel('x axis') # x label
plt.ylabel('y axis') # y label
plt.plot(x,y)

plt.subplot(1,3,2) #1 3  2
plt.title('<2>') #  
plt.plot(x,y2)

plt.subplot(1,3,3) #1 3  3
plt.title('<3>') #  
plt.plot(x,y3)

plt.show()
```
