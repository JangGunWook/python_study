```python
import numpy as np
```

```python
# np.genfromtxt : txt파일 불러오는 함수
# dtype
data = np.genfromtxt('./ratings.dat',delimiter = '::',dtype='int32')
data

# 실수 -> 과학적 표기법
# user_id, movie_id, rating, 시간(timestamp)
```

```python

print('배열 크기:', data.shape)
print('차원 수 : ', data.ndim)
print('전체요소 개수 : ', data.size)
```

```python
# 평점 데이터 접근
# 모든행, 열은 2
data[0:,2].mean()    
# 평균 구하는 함수 연결하기

```

```python
# 사용자 아이디만 접근
user_id= data[0:,0]
```

```python
# 사용자 아이디 중복없이 확인
user_id = np.unique(user_id)
print(user_id.size)
user_id
```

```python
# 사용자 아이디가 1인 데이터의 평점 평균
data[data[:,0]==1,2].mean()

data[np.where(data[:,0]==1),2].mean()
```

```python
# 5000 usr 평점평균
print(data[data[:,0]==5000])
data[data[:,0]==5000,2].mean()
```

```python
data[:,0]
```

```python


for id in user_id:
    sum = data[data[:,0]==id,2] # sum = data[data[:,0]==id][2]
    avg = np.mean(sum)
   
    print(f"{id}의 평균 평점 : {avg:.2f}")
```

```python
# 평균 평점이 4점 이상인 사용자는 몇명인가?
count = 0
for id in user_id:
    sum = data[data[:,0]==id,2] # sum = data[data[:,0]==id][2]
    avg = np.mean(sum)
    if  avg >= 4:
        count += 1
    
print(count)
    
```
