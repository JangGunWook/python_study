```python
# 판다스 가져오기
import pandas as pd
# 판다스를 불러와서 앞으로 pd라고 부르겠다!
```

### 판다스 객체 종류
- 1차원 : Series 시리즈 (인덱스 + 값)
- 2차원 : DataFrame 데이터 프레임(행,열+값/ 표형태)

#### 판다스 객체 다루기

```python
# 1차원 SEries 생성1: 리스트에 값을 입력하여 생성
num_series = pd.Series([3,4,5],index = ['son','kim','park'])
num_series
```

```python
# 1차원 Series 생성2 : 딕셔너리에 인덱스, 값 입력하여 생성
num_series2 = pd.Series({'son':3, 'Kim':4, 'park':5})
num_series2
```

```python
# 1차원 시리즈 속성확인
# numpy : 배열크기(.shap),차원의수(.ndim),데이터타입(dtype),전체요소(.size)
# 값, 인덱스, 데이터 타입, 배열크기
num_series2.values # 값만 확인
num_series2.index # 인덱스값만 확인
num_series2.dtype # 데이터 타입 확인
num_series2.shape # 데이터 크기확인
```

```python
# 1차원 시리즈 이름 지정
num_series2.name = 'name_cnt'
num_series2.name
```
