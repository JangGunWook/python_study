### Pandas 연산 및 유용한 함수

```python
# 파일 불러오기 및 저장
# exam = pd.read_csv('경로 및 파일명', 'encoding='euc-kr', index_col = 'id' )
# index_col : 해당 컬럼 값을 인덱스 값을 활용하는 키워드
exam = pd.read_csv('.\data\exam.csv')
exam
```

```python
# 데이터 파악하는 함수들
# 1.head() : 앞에서 5개 행 출력
exam.head()
# 2. tail() : 끝에서 4개 행 출력
exam.tail()
# 3. info() : 데이터 속성 정보 출력
exam.info()
# 4. shape : 행, 열 등의 배열의 크기 확인
exam.shape
# 5. describe() : 요약 통계량 확인
exam.describe()
```

```python
# 데이터 정렬하기
# 정렬 -> sort
# exam 데이터의 "수학" 컬럼 절령해 보기
#ascending = False : 내림차순
#ascending = True : 오름차순

display(exam['math'].sort_values(ascending = False) )
# exam.loc[:,'math']
# exam.iloc[:,2]

exam.sort_values(by ='math', ascending =False)

```

```python
# 총합을 구하는 함수 : sum()
# axis = 0, 행방향으로 데이터 확인 == 컬럼(열단위 처리)
# axis = 1, 열방향으로 데이터 확인 == 행(행단위 처리)
# 학생들의 수학, 영어, 과학 점수 총합 구하기

exam[['math','english','science']].sum()
exam.loc[:,['math','english','science']].sum()
exam.loc[:,'math':].sum(axis=1)
# 총합 컬럼 추가 : 파생변수(컬럼) 추가
exam['total'] = exam.iloc[:,2:].sum(axis = 1)
exam
```

```python
# 평균 구하는 함수 : mean()
# 각 학생의 3과목의 평균 구하기
exam['mean'] = exam.iloc[:,2:5].mean(axis = 1)
exam
```

```python
# 파생변수(컬럼) 추가2 - assign(컬럼 = 새로운 데이터())
# 실제 변수에 저장되는 것은 아님
# assign 함수 안에서 여러개의 컬럼을 생성할 수 있음
exam = exam.assign(mean2 = exam.iloc[:,2:5].mean(axis = 1))

# mean3 컴럼 삭제
exam.drop('mean2',axis = 1, inplace = True)
exam
```

```python
# np.where(조건,참,거짓)
import numpy as np
exam = exam.assign(result = np.where(exam['mean']>=60,'pass','fail'),
                   result1 = np.where(exam['mean']>=60,'pass','fail'),
                   result2 = np.where(exam['mean']>=60,'pass','fail'))

exam.drop(['result1', 'result2'], axis = 1, inplace =True)
exam
```
