```python
# 조건에 맞는 데이터 접근(불리언 인덱싱)
# 참 True, 거짓 False
# 불리언 인덱싱 예시 데이터 생성

sample_arr = [True, False]
np.random.seed(1)
bool_arr = np.random.choice(sample_arr, size = 10)
display(bool_arr)

tmp_arr = np.arange(0,10)
display(tmp_arr)

# 블리언 인덱싱
display(tmp_arr[bool_arr]) #True값만 나타난다
```

```python
# 블리언 인덱싱
# 성적을 담은 배열, 학생 번호를 담은 배열을 생성해 보자
# 50, 100까지 답은 배열 생성하기 1차원 배열로 요소값 20
np.random.seed(6)
big_score = np.random.randint(50,101, size = 20)
big_score

# 학생 번호 요소 20명
stu_num = np.arange(1,21)
display(stu_num)
display(big_score)
```

```python
# 논리연산자를 활용해서 big_score 배열 중 80이상인 성적 확인하기
big_score[big_score>= 80]
big_score >= 80

#70미만인 성적은 몇개일까?
display(big_score < 70)
# 요소 개수 세어 주는 키워드
big_score[big_score<70].size
```

```python
# 성적이 60점 이상인 학생 번호 출력하기
display(big_score >=60)
display(big_score[big_score >=60])
display(stu_num[big_score >=60])
display(stu_num[big_score >=60].size)

```

```python
data = np.loadtxt('./height_weight.txt',delimiter = ',')
data
```

```python
# 2)정보 확인하기
# 배열 크기, 데이터 타입, 차원 크기


print('배열의 크기 : ', data.shape)
print('배열의 크기 : ', data.dtype)
print('배열의 크기 : ', data.ndim)
```

```python
# 3) 키 변수에 키 데이터 10개 담아주기
height = data[0]
height 
```

```python
# 4) 몸무게 변수에 몸무게 데이터 10개 담아주기
weight = data[1]
weight
```

```python
# 5)키 데이터의 단위 변경하기 cm -> m(연산요소), 1m = 100cm
height = height *0.01
height
```

```python
# 6) 과체중만 출력하기
BMI = weight/(height*height)
display(BMI > 23)


과체중 = BMI[(BMI >=23) & (BMI <25) ] 
과체중
```
