### 데이터 접근 및 처리

```python
arr4 = np.array([10,20,30,40,50,60])
arr4
# 40요소 데이터에 접근하기
arr4[3]
# 10,30 데이터 동시에 접근
# tip : 여러 요소에 접근할 때는 대괄호 안에 대괄호를 한더 써줌
arr4[[0,2]]
```

```python
# 1차원 배열 여러 요소 접근하기(slicing)
arr4
# arr4[식작인덱스:끝인덱스+1]
#20~50 데이터 접근해보기
arr4[1:5]
```

```python
# 10~30 데이터 접근하기
arr4[0:3]
arr4[:3] #처음부터 슬라이싱이면 첫 인덱스 생략 가능
# 끝까지면 끝 인덱스 생략 가능

# 슬라이싱으로 10,30접근하기
arr4[:3:2]
```

```python
# 2차원 배열 접근
arr5 = np.array([[10,20,30],[40,50,60]])

# array 활용하기
arr5 = arr4.reshape((2,3))
arr5
```

```python
# 2차원 배열 인덱싱
# 2차원배열[행],[열] , 2차원배열[행,열]
# 30에 접근해복
arr5[0][2]
# arr5[행,열]
arr5[0,2]
```

```python
# 2차원 배열 슬라이싱
# 40 50 슬라이싱

arr5[1,0:2]
arr5[1,[0,1]]
```

```python
arr5[0:2,0:3:2]
arr5[0:2,[0,2]]
```
