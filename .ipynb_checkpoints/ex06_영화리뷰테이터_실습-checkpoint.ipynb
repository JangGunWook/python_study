```python
#pickle -> pkl
# [], () 자료구조를 python 형식 그대로 저장하게 하는 확장자 pkl
import pickle
# file = open('경로 및 파일명', '읽기 모드/ 쓰기 모드')
file = open("data\movie_review.pkl","rb")
data = pickle.load(file)
data #[리뷰문장내용, 긍정1/부정0]
file.close()
data
```

```python
# 1. 전체 리뷰 개수 확인하기
len(data)
```

```python
# 2. 리뷰 길이 평균값
sum = 0
for 길이 , 리뷰 in data:
    sum += len(길이)
      
    
avg = sum/len(data)
print(avg)
```

```python
# 3. 긍정리뷰 개수/ 부정리뷰 개수
# 긍정
posi = 0
# 부정
nega = 0

for 길이, 리뷰 in data:
    if 리뷰 == 1:
        posi +=1
    else:
        nega +=1

print(posi)
print(nega)
```

```python
#4. 리뷰 길이 최대값/리뷰내용
max = []
for 길이, 리뷰 in data:
   if len(max) < len(길이):
       max = 길이
       

print(len(max))       
print(max)
       
```

```python
# 5. 'ㅋ'이 포함된 리뷰 개수
cont = 0

for 리뷰, 평점 in data:
   if 리뷰.find('ㅋ') > -1:
       cont += 1
          
           
print(cont)            
```

```python
# 6. '검색데이터'이 포함된 리뷰 개수
cont = 0
Que = input("검색하고자 하는 데이터 >>")

for 리뷰, 평점 in data:
   if 리뷰.find(Que) > -1:
       cont += 1
          
           
print(cont)      
```
