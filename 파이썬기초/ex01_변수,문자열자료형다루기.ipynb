### 문자열 접근하는 방법
 - 문자열 -> 데이터
 - 인덱스, 슬라이싱

```python
s = "My name is JY"
print(s[0])
print(s[8])
print(s[-2])
print(s[-1])
```

```python
# 특정문자열에 대해서 접근하게 되는 개념
# [시작인덱스 : 끝인덱스 + 1]
# MY를 슬라이싱

s[0:2]

# name
print(s[3:7])

# is
print(s[8:10])
```

```python
# 생략할 수 있는 인덱스 번소
# s 변수에서 맨 첫번째부터 name의 e까지 실라이싱

print(s[0:7])
print(s[:7]) #0번부터 출력할거라면 생략가능

print(s[8:13])
print(s[8:]); #맨마지막까지 출력할거라면 생략가능
print(s[:]); #모든 문자열에 접근
print(s[::2]) #모든 문자열의 2칸씩 출력
```

```python
day = "2020년 3월 3일의 날씨는 맑음입니다."

print('날짜 :', day[:11])
print('날짜 :',day[-6:-4])
```

```python
# 문자열 관련된 함수
st = day.find('2020') #해당 단어가 어디부터 시작되어 있는지 출력
ed = day.find('3일')+2 #9 +2

print(st, ed)
day[st:ed]
```

### 문자열관련 함수 정리

```python
# 길이 확인하는 함수
len(now)
now[10:len(now)]
```

```python
# 문자열 대소문자 변환 함수

# 소문자 ---> 대문자 바꾸기
eng_str = 'hi,My name is JY'
eng_str1 = eng_str.upper() # 해당 변수에 초기화X

# 대문자 ---> 소문자 바꾸기
eng_str3 = eng_str.lower()
eng_str3

```

```python
# '찾을 문자열'이 몇개 있는지 확인하는 함수
# 'm' 카운트
eng_str3.count('m') # 이문장에서 m은 두번 들어있다!!
eng_str3.count('M') # 이문장에서 M은 두번 들어있다!!
eng_str.count('m')
eng_str.lower().count('m')
```

```python
# 특정 문자열 찾아서 인덱스 번호 확인(반환)하는 함수
# find()
# index()

eng_str.find('i') #제일 먼저 찾아지는 'i'인덱스 값 출력
eng_str.index('i') #제일 먼저 찾아지는 'i'인덱스 값 출력

# 문자열에 없는 단어 찾아달라고 하기
eng_str.find('z') #해당 단어가 없으면 -1값 반환
eng_str.index('z') #해당 단어가 없으면 error가 발생
```

```python
# strip() : 문자열 양끝의 공백을 제거하는 함수

"    배고프다!      ".strip()
```

```python
# 문자열1을 문자열2로 변경하는 함수
# replace(문자열1, 문자열2)

print(eng_str)
eng_str.replace("JY", "JH")
```

```python

article = "[요기요, 화이트데이 맞이 '선물하기' 이벤트 진행] 기사입력 2022.03.10. 작성자 이나영기자 요기요는 화이트데이를 앞두고 요기요 선물하기 서비스를 이용하는 고객들을 위한‘너와 나의 선물고리’이벤트를 진행한다고 10일 밝혔다."
start = article.find("2022")
end = article.find(".10")+4
day = article[start:end]
print(start)
print(end)
print(day)
print(f"기사가 업로드된 일자는{day}입니다")
```
