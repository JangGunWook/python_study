```python
score_dict = {'이름':['손지영','김은영','김운비','채수민','최성우'],
        'Python':[90,100,85,90,80],
        'Java':[85,80,100,95,85],
        'html/css':[75,70,90,80,90]}
score_dict

sum2 = 0
score_dic_key = list(score_dict.keys())
score_dic_key = score_dic_key[1:]

name = input("이름을 정확하게 기입하세요 >> ")

name_list =list(score_dict['이름'])

if name in name_list:
    indnum = name_list.index(name)
    for i in score_dic_key:
        sum2 += score_dict[i][indnum]
    print(sum2/3)
    

else:
    print("이름이 잘못 되었습니다")

```

```python
s_sum = 0
name = input('이름을 정확하게 입력하세요 >')

for i in range(0,len(score_dict['이름'])):
    if(score_dict['이름'][i] == name):
        s_sum += score_dict['Python'][i]
        s_sum += score_dict['Java'][i]
        s_sum += score_dict['html/css'][i]
        print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')
```

```python
_sum = 0
name = input('이름을 정확하게 입력>> ')
# 유무를 체크하는 변수 
check = False

for i in range(0,len(score_dict['이름'])):
    if (score_dict['이름'][i] == name):
        check = True # 정보가 있었음 , 이름이 찾아졌음
        s_sum += score_dict['Python'][i]
        s_sum += score_dict['Java'][i]
        s_sum += score_dict['html/css'][i]
        print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')

if not check: 
    print('검색 결과가 없습니다.')
```
