```python
# 리스트 선언하기
a = [] #빈리스트
b = [1,2,3] #숫자 데이터를 담은 리스트
c = ['MY','name','is','JY'] # 문자열 데이터를 담은 리스트
d = [ 1, 2 , 'My','name'] # 숫자, 문자열 데이터를 담은 리스트
e = [1, 2, ['My', 'name']] # 이중 리스트 가능

print(a)
print(b)
print(c)
print(d)
print(e)
print(type(e)) #클래스는 list다

```

#### 리스트 데이터 접근
 - 인덱스 번호 활용
 - 문자열 인덱싱, 슬라이싱 활용과 동일
 - 인덱싱 : 무엇인가를 '가리킴'
 - 슬라이싱 : 무엇인가를 '잘라냄'(범위)
 - 대상[인덱스 번호], 리스트[시작:끝+1]

```python
list1 = [2,5,7,9,10]
# 2접근
list1[0]
# 9접근
list1[-2]
```

```python
list2 = [1,2,3,['a','b','c']]
temp = list2[3]
print(temp)
list2[3][1]
```

```python
# 슬라이싱
# 리스트[시작:끝+1]
list3 = ['지수', '문규','다인','수양','성문']
list3[0:2]

# 슬라이싱 결과는 여러개에 대해 리스로 묶어서 반환
```

```python
#'다인','성문' 슬라이싱 해보기
print(list3[2::2])
```

```python
# 맨 뒤에부터 앞에까지 차례대로 출력하기
list3[::-1]
```

```python
# 리스트 가변적이다!
# 리스트 더하기
list4 = ['인우', '지웅', '강련','서준희']
print(list4 + list3)
```

```python
# 리스트 곱하기
list5 = [1,2,3]
print(list5*3) # 수학적 개념의 연산X, 자료의 크기를 그만큼 늘림
```

```python
# 리스트 맨 뒤쪽 값 추가
#  리스트.append(값)
list5.append(4) # appand(자료하나만 가능)
```

```python
lst = ["딸기", "바나나", "수박", "체리", "포도"]

Question = input("검색할 문자를 입력하세요> ")
if Question in lst:
    print(f"{Question}는 리스트에 {lst.index(Question)}번재 인덱스에 들어있습니다")
else:
    print(f"{Question}는 리스트에 드러있지 않습니다")
```

```python
s = "Hi, My name is Mihee"


Question = input("검색할 문자를 입력하세요 : ")

if Question in s:
    print(f'\'{Question}\'는 {s.count(Question)}번 들어 있습니다')
else:
    print(f"{Question}는 문자열에 들어 있지 않습니다")
        
```
