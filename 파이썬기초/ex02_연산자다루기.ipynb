### 산술연산자

```python
# 나누기, 나머지, 나누기(몫) 구하기
# num1 변수 10 대입
# num2 변수 7 대입
num1 = 10
num2 = 7
print("나누기 결과 : " , num1/num2)
print("나누기 몫 결과 : " , num1//num2)
print("나누기 나머지 결과 : " , num1%num2)
```

```python
# 문자열 더하기
#  str1 "안녕" 대입하기
#  str2 "하세요" 대입하기
#  두개의 변수 더하기 출력
str1 = "안녕"
str2 = "하세요"
print(str1 + str2)
```

```python
str3 = '10'
str4 = '7'
print(str3 + str4)
```

```python
# num1 과 str4는 더해질까?
# 숫자 + 문자열 더해지지 않음

# 강제형변환
# python 형변환 int(대상),str(대상)
str(num1) + str4
num1 + int(str4)
```

### 복합연산자

```python
#  = 대입연산자
#  += 더하기 대입
#  -= 빼고 대임
# *= 곱하고 대임
#  /= 나누기 대입
#  %= 나머지값 대입

#  += 기준으로 으로 설명
a = 3
b = 2
a += b
print(a)
```

### 비교연산자

```python
#  a>b : a가 b보다 크다
#  a >= b : a가 b보다 크거나 같다
#  a < b : b가 a보다 크다
#  a <= b : b가 a보다 크거나 같다
#  a==b : a와b가 같다
#  a!=b : a와b는 다르다(같지 않다)
```

### 난수생성(random)

```python
# random 모듈 라이브러리
import random
a = random.randint(1,10)
b = random.randint(1,10)
print(a<b)
print(b>a)
print(b==a)
print(b!=a)
print(b,a)
```

### 논리연산자

```python
# and : 논리1 and 논리2
# a>b and a!=b     ==========>  true
# a<b and a!=b     ==========>  false

# or : 논리1 or 논리2
# true or false    ==========> true
# true or true     ==========> true

# not : not 논리
# 논리를 뒤집는 것
```

```python
print(2>1 and 2==1)
print(True or False)
print(not False)  # if 조건식 부분에서 많이 사용
```

### 멤버연산자

```python
# 조건문에 자주 활용이 됨
# in : 찾고자 하는 값(x)이 포함되어 있으면 true
# not in : 찾고자 하는(x) 값이 포함되지 않으면 true

#  x in 문자열
'a' in 'aaaaaaabbbbbbbbbbb' # 뒷 문장안에 a가 있어? true

#  x not in 문자열
'a' not in 'aaaaabbbbbbbbbbb'  # 뒷 문장안에 a가 없어? false
```

```python
'a' in ['a','b','c','d'] 
```

### 삼항연산자

```python
# a if 조건식 else b
score = 80 # 60점 이상
'합격' if score >= 60 else '불합격'
```

```python
num = int(input("정수 입력 >> "))
result = "홀수" if num%2==1 else "짝수"
print(num, "는(은) " + result + "입니다")
```
