## 이중 리스트 반복문 활용하기

```python
frd_list = [['강련',28],['건욱',29],['선웅',26],['주섭',27],['나행',25]]
print('리스트의 길이는 ?', len(frd_list))
```

```python
# 모든 사람의 나이 데이터만 출력하시오
# 반복문 for

for i in range(0,len(frd_list)):
    print(frd_list[i][1])
```

```python
# frd_list에 들어있는 사람의 이름을 구하기!
name = input("이름을 입력하시오! >> ")

for i in range(0,len(frd_list)):
    if frd_list[i][0] == name:
        print(f"{name}님의 나이는 {frd_list[i][1]}")
```

```python
print(frd_list[1])
```

```python
for name, age in frd_list:
    print(name, age)
```

```python

# 이름과 나이를 담는 리스트를 만들어보기
name_list = []
age_list = []

for name, age in frd_list:
    name_list.append(name)
    age_list.append(age)
    
print(name_list)
print(age_list)
```
